```python
from bs4 import BeautifulSoup
import requests
```

```python
def changeUrl(url):
    sub_url=url
    page=requests.get(sub_url)
    soup=BeautifulSoup(page.content,"html.parser")
    lists=soup.find_all("div",class_='d6e81fd0')
    print(url,page)
    return lists
```

```python
main_url="https://www.bproperty.com/en/bangladesh/properties-for-rent/"
lists=changeUrl(main_url)
title_l=[]
location_l=[]
price_l=[]
type_l=[]
bed_l=[]
bath_l=[]
area_l=[]
```

```python
def aria_label_valid(label):
    if label==None:
        data="NuLL"
    else:
        data=label.text.replace('\n', '').strip()
    return data
```

```python
def makeList(lists):
    for list in lists:
        title = aria_label_valid(list.find(attrs={'aria-label': 'Title'}))
        price = aria_label_valid(list.find(attrs={'aria-label': 'Price'}))
        location = aria_label_valid(list.find(attrs={'aria-label': 'Location'}))
        type = aria_label_valid(list.find(attrs={'aria-label': 'Type'}))
        beds = aria_label_valid(list.find(attrs={'aria-label': 'Beds'}))
        baths = aria_label_valid(list.find(attrs={'aria-label': 'Baths'}))
        area = aria_label_valid(list.find(attrs={'aria-label': 'Area'}))  
        title_l.append(title)
        location_l.append(location)
        price_l.append(price)
        type_l.append(type)
        bed_l.append(beds)
        bath_l.append(baths)
        area_l.append(area)


```

```python
for pageNumber in range(1,720,1):
    if pageNumber==1:
        makeList(lists)
    else:
        lists=changeUrl(f"{main_url}page-{pageNumber}/")
        # lists=changeUrl("https://www.bproperty.com/en/bangladesh/properties-for-rent/page-15/")
        makeList(lists)
# print(infos)
```

```python
dict={"title":title_l,
      "location":location_l,
      "type":type_l,
      "bed":bed_l,
      "bath":bath_l,
      "area":area_l,
      "price":price_l}
```

```python
import pandas as pd
    
df = pd.DataFrame(dict)
```

```python
df.to_csv('house_rent.csv',index=False)
```
